```python
import pandas as pd

# Read your CSV data file into a DataFrame
acs_df = pd.read_csv('/Users/mahshid/Downloads/acs2017_census_tract_data.csv')

# Step 1: Filter out unneeded columns
columns_to_drop = ['Men', 'Women', 'Hispanic', 'White', 'Black', 'Native', 'Asian', 
                   'Pacific', 'VotingAgeCitizen', 'ChildPoverty', 'Professional', 
                   'Service', 'Office', 'Construction', 'Production', 'Drive', 
                   'Carpool', 'Transit', 'Walk', 'OtherTransp']
acs_df = acs_df.drop(columns=columns_to_drop)

# Step 2: Aggregate the data

data_columns = ['TotalPop', 'IncomePerCap', 'Poverty']  
county_info = acs_df.groupby(['State', 'County'])[data_columns].sum().reset_index()

# Step 3: Add an ID column for the county
county_info['County_ID'] = county_info.index + 1

# Display the DataFrame
pd.set_option('display.max_rows', 100)
county_info


```

```python

import pandas as pd

# Read your CSV data file into a DataFrame
acs_data = pd.read_csv('/Users/mahshid/Downloads/acs2017_census_tract_data.csv')

# Define a function for weighted mean
def weighted_mean(x, data, weight):
    d = data.loc[x.index, weight]
    return (x * d).sum() / d.sum()

# Create a dictionary that maps columns to the operations you want to perform on them
ops = {'TotalPop': 'sum', 
       'IncomePerCap': lambda x: weighted_mean(x, acs_data, 'TotalPop'), 
       'Poverty': lambda x: weighted_mean(x, acs_data, 'TotalPop')}

# Now group by 'County' and 'State', and calculate using the operations in the dictionary
county_info = acs_data.groupby(['County', 'State']).agg(ops).reset_index()

# Add an ID column
county_info['ID'] = range(1, len(county_info) + 1)

# Now when you query, you should get one row per county per state
counties = [("Loudoun County", "Virginia"), 
            ("Washington County", "Oregon"), 
            ("Harlan County", "Kentucky"), 
            ("Malheur County", "Oregon")]
results = pd.DataFrame()

for county, state in counties:
    data = county_info.loc[(county_info['County'] == county) & (county_info['State'] == state), 
                           ['ID', 'County', 'State', 'TotalPop', 'Poverty', 'IncomePerCap']]
    results = pd.concat([results, data])

# Reset the index of the results DataFrame
results.reset_index(drop=True, inplace=True)

# Display the results
results









```

```python
# Get the row with the maximum 'TotalPop' value
most_populous_county = county_info.loc[county_info['TotalPop'].idxmax()]

# Get the row with the minimum 'TotalPop' value
least_populous_county = county_info.loc[county_info['TotalPop'].idxmin()]

# Display the results
print("Most populous county in the USA:")
print(most_populous_county)
print("\nLeast populous county in the USA:")
print(least_populous_county)
```

```python
import pandas as pd

# Load the data
County_info = pd.read_csv('/Users/mahshid/Downloads/acs2017_census_tract_data.csv')
covid_data = pd.read_csv('/Users/mahshid/Downloads/COVID_county_data.csv')

# Create a 'County Name' column in each dataframe
County_info['County Name'] = County_info['County'] + ', ' + County_info['State']
covid_data['County Name'] = covid_data['county'] + ', ' + covid_data['state']

# Convert the date to a datetime format
covid_data['date'] = pd.to_datetime(covid_data['date'])

# Create year and month columns
covid_data['year'] = covid_data['date'].dt.year
covid_data['month'] = covid_data['date'].dt.month

# Merge the dataframes
merged_data = pd.merge(covid_data, County_info, on='County Name', how='inner')

# Filter the data for Malheur County, Oregon in specific months
desired_rows = [
    ('Malheur County, Oregon', 2020, 8),
    ('Malheur County, Oregon', 2021, 1),
    ('Malheur County, Oregon', 2021, 2)
]
filtered_data = merged_data[
    merged_data[['County Name', 'year', 'month']].isin(desired_rows).all(axis=1)
]

# Select the required columns
desired_columns = ['County Name', 'month', 'cases', 'deaths']
filtered_data = filtered_data[desired_columns]


# Reset the index
filtered_data = filtered_data.reset_index(drop=True)

# Display the resulting table
print(filtered_data)

```
