```python
import pandas as pd
from datetime import datetime, timedelta

# Function to decode TriMet's timestamp representation
def create_timestamp(row):
    opd_date = datetime.strptime(row['OPD_DATE'], '%d%b%Y:%H:%M:%S')
    act_time = timedelta(seconds=row['ACT_TIME'])
    timestamp = opd_date + act_time
    return timestamp

# Load data
data = pd.read_csv('/Users/mahshid/Downloads/bc_veh4223_230215.csv')

# Drop unwanted columns
data = data.drop(columns=['EVENT_NO_STOP', 'GPS_SATELLITES', 'GPS_HDOP'])

# Create the TIMESTAMP column
data['TIMESTAMP'] = data.apply(create_timestamp, axis=1)

# Display the DataFrame
data.head()




```

```python
data.head()
```

```python
# Calculate the difference in METERS and TIMESTAMP for each row
data['dMETERS'] = data['METERS'].diff()
data['dTIMESTAMP'] = data['TIMESTAMP'].diff().dt.total_seconds()

# Calculate SPEED = dMETERS / dTIMESTAMP
data['SPEED'] = data.apply(lambda row: row['dMETERS'] / row['dTIMESTAMP'] if row['dTIMESTAMP'] != 0 else 0, axis=1)

# Drop the unneeded dMETERS and dTIMESTAMP columns
data = data.drop(columns=['dMETERS', 'dTIMESTAMP'])

# Display the DataFrame
data.head()
```

```python
# Remove rows with negative SPEED values
filtered_data = data[data['SPEED'] >= 0]

# Calculate the statistics for the SPEED column
filtered_data['SPEED'].describe()
```

```python
# Maximum speed
max_speed = filtered_data['SPEED'].max()

# Location and time of maximum speed
max_speed_info = filtered_data.loc[filtered_data['SPEED'].idxmax()]

# Median speed
median_speed = filtered_data['SPEED'].describe()['50%']

print(f"Maximum speed for vehicle #4223 on February 15, 2023: {max_speed} m/s")
print(f"Location of maximum speed: (latitude: {max_speed_info['GPS_LATITUDE']}, longitude: {max_speed_info['GPS_LONGITUDE']})")
print(f"Time of maximum speed: {max_speed_info['TIMESTAMP']}")
print(f"Median speed for vehicle #4223 on February 15, 2023: {median_speed} m/s")
```
